```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('extracted.csv')
df = df.drop(['Unnamed: 0','Unnamed: 1'], axis = 1)
```

```python
df
```

```python
df['Label'] = 'Episodic'
```

```python
transform = df[['Label','Episodic_Z']].rename(columns={'Episodic_Z':'data'})
transform
```

```python
composite = df[['Label','Cog_Domain_Z_Avg']].rename(columns={'Cog_Domain_Z_Avg':'data'})
composite['Label'] = 'Composite'
semantic = df[['Label','Semantic_Z']].rename(columns={'Semantic_Z':'data'})
semantic['Label'] = 'Semantic'
executive = df[['Label','Executive_Z']].rename(columns={'Executive_Z':'data'})
executive['Label'] = 'Executive'
vs = df[['Label','Visuospatial_Z']].rename(columns={'Visuospatial_Z':'data'})
vs['Label'] = 'VS'
attention =df[['Label','Attention_Z']].rename(columns={'Attention_Z':'data'})
attention['Label'] = 'Attention'
```

```python
transform = pd.concat([composite, transform, semantic,executive,vs,attention], axis=0).reset_index(drop=True)
```

```python
transform.Label.unique()
```

```python
transform = transform.rename(columns={'data':'Score'})
```

```python
sns.swarmplot(data = transform, x='Label',
              y='Score',
              )
sns.violinplot(x="Label", y="Score", data=transform, inner=None, palette=['w'])
```

```python
fig, ax = plt.subplots(figsize=(6.5,4.5))

#ax = sns.swarmplot(data=transform, x='Label',
#              y='Score',
 #             palette=['black','white', 'blue', 'purple','green','gold','brown'],
  #                 order=('Composite', '', 'Episodic', 'Semantic', 'Executive', 'VS', 'Attention')
  #            )


    
ax = sns.violinplot(x="Label",
                    y="Score",
                    data=transform,
                    inner = None,
                    scale='width',
                    palette=['black','white', 'blue', 'purple','green','yellow','red'],
                   order=('Composite', '', 'Episodic', 'Semantic', 'Executive', 'VS', 'Attention'))

ax = sns.boxenplot(data=transform, x='Label',
             y='Score',
            color = 'white',
            width = 0.1,
           palette=['white','white', 'white', 'white','white','white','white'],
           order=('Composite', '', 'Episodic', 'Semantic', 'Executive', 'VS', 'Attention')
           )


#y ticks labels
plt.yticks(size='medium')
ax.yaxis.set_tick_params(direction='in', length=5) # Set the tickmarks to point outward

# set xlabel
plt.text(3.2, -3, 'Cognitive Domain',weight='bold', size = 'x-large')
ax.xaxis.set_tick_params(direction='in', length=5)

# set ylabel
plt.ylabel('Z Score',weight='bold', size="xx-large")

# set xticks
plt.xticks()
ax.axes.get_xaxis().set_visible(False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig('Cog_Score', bbox_inches='tight', dpi = 500)
```
